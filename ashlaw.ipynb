```python
# Simple working LightRAG test with async initialization
import sys
import asyncio
import nest_asyncio
sys.path.append('.')
from pathlib import Path

# Apply nest_asyncio for Jupyter
nest_asyncio.apply()

async def test_lightrag():
    print("ðŸ” Testing LightRAG with async initialization...")
    
    # Setup LightRAG the simple way
    from lightrag import LightRAG
    from lightrag.utils import EmbeddingFunc
    from lightrag.kg.shared_storage import initialize_pipeline_status
    import google.generativeai as genai
    from sentence_transformers import SentenceTransformer
    import numpy as np
    
    GOOGLE_API_KEY = "AIzaSyC7PyJd1PWkVT9Wk95GphulPb_ztonEbQI"  # Replace
    
    # Simple LLM function
    async def llm_func(prompt, **kwargs):
        genai.configure(api_key=GOOGLE_API_KEY)
        model = genai.GenerativeModel("models/gemma-3-27b-it")
        response = model.generate_content(prompt)
        return response.text
    
    # Simple embedding function  
    async def embed_func(texts):
        model = SentenceTransformer("all-MiniLM-L6-v2")
        return model.encode(texts, convert_to_numpy=True)
    
    # Create LightRAG
    rag = LightRAG(
        working_dir="./data/lightrag",
        llm_model_func=llm_func,
        embedding_func=EmbeddingFunc(
            embedding_dim=384,
            max_token_size=8192,
            func=embed_func
        )
    )
    
    # Initialize storage
    await rag.initialize_storages()
    await initialize_pipeline_status()
    
    print("âœ… LightRAG initialized successfully")
    
    # Test with your markdown files
    markdown_dir = Path("./data/markdown_files")
    markdown_files = list(markdown_dir.glob("*.md"))
    
    if markdown_files:
        # Read first file
        with open(markdown_files[0], 'r', encoding='utf-8') as f:
            content = f.read()
        
        # Insert document
        rag.insert(content)
        print(f"âœ… Inserted document: {markdown_files[0].name}")
        
        # Test query
        from lightrag import QueryParam
        result = rag.query("Extract entities", param=QueryParam(mode="mix"))
        print(f"âœ… Query result length: {len(str(result))}")
        print(f"ðŸ“ First 150 chars: {str(result)[:150]}...")
        
    return rag

# Run the test
rag = asyncio.run(test_lightrag())
print("ðŸŽ‰ LightRAG working!")
```
